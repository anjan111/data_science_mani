```python
# To create the  Model
import numpy as np
from math import sqrt
import matplotlib.pyplot as plt
%matplotlib inline
def simple_linear_regression(X_train , y_train):
        x = X_train
        y = y_train
        N = np.size(x)
        x_m = np.mean(x) # X mean
        y_m  =np.mean(y) # y  mean      
        B1 =  (np.sum(x*y) - N*x_m*y_m)   / (np.sum(x*x) - N*x_m*x_m)        
        B0 =  y_m  -  B1  * x_m
        return (B0,B1)

# TO validate the Model   actual  ----> y_text ,Predicted --->y_pred
def root_mse(y_test , y_pred):
    sum_square_error = 0
    for i in range(len(y_test)):
        error = y_test[i]-y_pred[i]
        sum_square_error += error ** 2
    mean_sum_square_error = sum_square_error / float(len(y_test))
    rmse = sqrt(mean_sum_square_error)    
    return rmse

# Finding the prediction
def prediction(B0,B1 , X_test):
    y_pred = B0 + B1 * X_test
    return y_pred

def data_visua(X_train ,y_train,X_test,y_test,y_pred):
    # data 
    plt.scatter(X_train,y_train)
    plt.xlabel("X_train")
    plt.ylabel("y_train")
    plt.show()
    # model visualization
    plt.plot(X_test,y_pred)
    plt.xlabel("X_test")
    plt.ylabel("y_pred")
    plt.show()
    # error term
    len1 = np.arange(len(X_test))
    print len1
    print y_test
    plt.plot(len1 ,y_test - y_pred)
    plt.xlabel("numbers")
    plt.ylabel("error term")
    plt.show()
def main():
    X_train =np.array([0,1,2,3,5,7,9]) 
    y_train =np.array([1,2,4,6,10,14,18])
    X_test  = np.array([4,6,8])
    y_test  = np.array([8,12,16])
    B       =  simple_linear_regression(X_train , y_train)  # Y = B0 + B1 * X
    B0 = B[0]
    B1 = B[1]
    print"Simple Linear Regression"
    print "Y = ",B0 ," + ",B1," * X"
    y_pred  =  prediction(B0,B1 , X_test)
    rmse    =  root_mse(y_test , y_pred)

    print  "Final Simple Linear Regression"
    print "Y = ",B0 ," + ",B1," * X +/-", rmse
    data_visua(X_train ,y_train,X_test,y_test,y_pred)
if __name__ =="__main__":
    main()
```
